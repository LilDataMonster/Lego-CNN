````python
import os
import json
import pandas as pd

from detector.mrcnn import lego
````

````python
ROOT_DIR = './'
os.chdir(ROOT_DIR)
````

````python
df = pd.read_csv('protos/hex_colors.txt', header=None)
df
````

````python
# get dataset class names
DATASET = "val"
LEGO_DIR = os.path.join(ROOT_DIR, "datasets", "output", "coco_data")

dataset = lego.LegoDataset()
dataset.load_coco(LEGO_DIR, DATASET)
dataset.prepare()
print("Images: {}\nClasses: {}".format(len(dataset.image_ids), dataset.class_names))
````

# CVAT Task/Project Constructor

````python
output = []
for _id, (_name, color) in enumerate(zip(dataset.class_names, df.values)):
    output.append({"id": _id, "name": _name, "color": color[0], "attributes": []})
print(json.dumps(output, indent=2, sort_keys=True))
````

# CVAT Auto Annotation Constructor

````python
# cvat output
output = []
for _id, _name in enumerate(dataset.class_names):
    output.append({"id": _id, "name": _name})
print(json.dumps(output, indent=2, sort_keys=True))
````

# Protos

````python
from google.protobuf import text_format
from protos.coco_label_map_pb2 import StringIntLabelMapItem, StringIntLabelMap
````

````python
# protobuf output
output_label_items= []
for _id, _name in enumerate(dataset.class_names):
    output_label_items.append(StringIntLabelMapItem(name=_name, display_name=_name, id=_id))

label_map = StringIntLabelMap(item=output_label_items)

text_proto = text_format.MessageToString(label_map)
print(text_proto)
````

````python
with open("lego_label_map.pbtxt", "w") as pbfile:
    text_format.PrintMessage(label_map, pbfile)
````
